## Introduction

### The separator provides a place for any liquid to deposit from the gas. The separator pressure is set above the pipeline pressure so that the gas can go through the necessary cooling, treating, dehydration, and gas processing - each with some pressure drop - and reach the required pipeline pressure. The gas-liquid separation process usually occurs in a separation tank into which the well fluid flows allowing the gas, oil, and water to separate due to the action of gravity. A separator can be two-phase, separating the gas from the liquids, or three-phase, separating the gas, oil, and water. The Figure  below shows two types of three-phase gravity separation tanks: horizontal and vertical. The main problem is the size of the separator tanks used today. To achieve a large separation capacity the tanks need to be of large dimensions, which increases the weight and space occupied and consequently increases the cost of the processing facilities.
### The separation process can be accelerated using centrifugal separation, which can have an effect a hundred times greater than gravity (SLOT, 2013). This type of equipment is known as a cyclone. In this case, the less dense phase will be forced by the centrifugal field to migrate toward the center while the more dense phase is forced to move toward the wall. Subsequent to segregation, the phases can be extracted by different extraction devices.

#### Figure 1: Three-phase separation tanks: (a) horizontal and (b) vertical (Aong, 2020).


## Types of Hydrocyclone

###  Inlet of the flow: There is the tangential inlet flow, where the flow enters through one or more tubes connected tangentially to the cyclone, in case of multiple inlets, they are distributed equidistantly on the circumference. In the axial inlet, we usually have a static element with vanes that accelerates and imposes rotation components to the flow. In the picture 2 illustrates the two forms of cyclones according to their input.
###  Tube geometry: In the conical form there is a reduction in the diameter of the separation tube. In the cylindrical shape, there is no reduction in the diameter of the separation tube, generally used in compact cyclones of axial application.
###  Form of extraction: there is the counter-current configuration where the heavy phase exits downstream of the flow and the light phase upstream of the flow, usually found in tangential inlet cyclones. The other alternative is co-current as both exits arepositioned downstream of the flow.

#### Figure 2: Tangential inlet cyclones (HOFFMANN; STEIN, 2008) and (b) axial inlet(TYVOLD, 2015).



## Axial Hydrocyclone

### This type of cyclone is more compact when compared to the tangential inlet cyclone and takes advantage of being able to be installed "in-line"with the production pipe, as shown in the figure 3. This feature has made this device recognized as an "inline separator".
### Besides being compact, the axial inlet cyclone is static, which lowers maintenance costs, leads to a lower level of instrumentation, control and monitoring. The reduction in operating costs extends the period when production is economically favorable providing increased oil and gas recovery and the economic potential of the field. A disadvantage of the axial inlet cyclone is the possibility of bubble breakage due to high shear stresses and turbulence effects, which hinders phase segregation.
### Turbulent flow with swirl is essentially three-dimensional and transient leading to often complex numerical modeling, raising the computational cost. Using turbulent viscosity models to express Reynolds stresses in terms of averaged quantities is not suitable for high-swing flow, where the current lines are curved (POPE, 2000).

#### Figure 3: Axial inlet cyclonic separator (SAVREE, 2020).


## Motivation of study

### The main motivation of this work is the wide use of cyclones in different areas of engineering, aiming the separation of phases. Specifically, the gas-liquid separation is important because of the pre-salt reserves that have light oils with large amounts of CO2. It is of interest to oil and gas operators to have a virtual laboratory, obtainable through numerical simulations, that will allow them to refine and improve the efficiency of this type of separator.
### In this study, attention is focused on the turbulent flow with swirl induced from fixed vane swirl generating element (SGE). Since the flow is simplified to a two-dimensional model, the SGE is not part of the computational domain. Considering a dispersed flow where the gas is the dispersed component and the oil the continuous phase, one will have the gas flowing through the center of the pipe and the oil near the wall. Finally, it is intended to obtain the optimal tail geometry.

## Objective

The velocity field (axial, radial and tangential components), the pressure field and the swirl decay due to the viscous effect are investigated.

## Axial Velocity - Discussion

### The relationship between axial and tangential velocity is directly linked to a slewing flow. As a way to evaluate this relationship, figure below presents the radial distribution of the axial velocity of the three cases, for the 5D, 20D, 40D and 60D sections. It is known that the axial flow inlet velocity is the same for all cases and that the tangential velocity increases the greater the deflection angle. Thus, the influence of the tangential component on the axial velocity is noted, since the uniform axial velocity on the inlet is the same for all cases, but there is a gradual difference between the three deflection angles, where for the case that has an angle of 72 there is a maximum speed greater than 2 m/s. The axial velocity profile concentrated close to the wall is related to the geometric shape of the flow inlet, which is delimited by the annular between the tail and the tube wall. As the flow develops, the four analyzed sections present similar patterns for all cases. Near the centerline there is an accentuation of negative axial velocity between the 5D and 20D sections. This identifies the flow recirculation region where the vortices are formed. With the development of the flow, there is an increase in the axial velocity in the center and a decrease near the wall, which tends to conserve the amount of movement and reach a fully developed flow regime.

## Tangencial Velocity - Discussion

### The tangential velocity distribution is shown plotted in the figure below. There are three well-defined regions, according to the division presented in the literature (KITOH, 1991), for all studied cases. Near the wall there is a thin boundary layer, starting from the wall towards the centerline there is a region where the tangential velocity progressively increases until reaching a maximum peak, this region is called free vortex or potential. The third and last region comprising the area between the centerline and the tangential velocity peak is called the solid body vortex. These three regions are similar to the rankine vortex close to the inlet, as the structure tends to change as it moves away from the input due to the gyro dissipation due to viscosity. The wall velocity does not reach zero as the mesh is not refined enough to fully solve the boundary layer, so the wall functions for y+ are necessary, Slot (2010) and Karimpoorheidari (2019) also noted this same behavior.

## Swirl number - Code

```python
def plotS(df, distances, angle):
  
  Sss = []
  for d in distances:
    Sss.append ( (2*np.pi*(np.trapz((df['dados_radial']['vel_axial_' + str(int(d)) + 'd']*df['dados_radial']['r_'+ str(int(d)) +'d']*df['dados_radial']['vel_tang_'+ str(int(d)) +'d']*df['dados_radial']['r_'+ str(int(d)) +'d']), df['dados_radial']['r_'+ str(int(d)) +'d'])))/(np.pi*(R**3)*(Ub**2)))
  S_tot = Sss #S_5d,S_10d,S_15d,S_20d,S_30d,S_40d,S_50d,S_60d
  x = distances

  #Grfico Swirl Number
  plt.scatter(x, S_tot,marker='o', label='Numerical '+str(int(angle))+'')
  
  plt.rcParams['xtick.labelsize'] = 16
  plt.rcParams['ytick.labelsize'] = 16
  plt.xlabel("x/D", fontdict={'size': 20})
  plt.ylabel("S/$S_{0}$", fontdict={'size': 20})
  plt.grid(True)
  plt.legend(loc="best", prop={'size': 15})
  #plt.savefig('graficos/intensidade_giro.png')
  plt.show;
```

## Swirl Number - Discussion

### The operating efficiency is directly related to the centrifugal force generated downstream of the vanes. For liquid-liquid separation, one of the challenges is to overcome the low separation efficiency due to the low density difference between the phases (SLOT, 2010). Therefore, some projects seek to overcome this challenge. Although this study is single-phase, the analysis is carried out for different deflection angles, analyzing the behavior and pattern of the flow. The rotational intensity, given by the equation x in the modeling section, is crucial for this analysis. In the figure below, there is the decay of the rotation intensity, as observed by Ddirkzwager (1996), Slot (2010) and Karimpoorheidari (2019) AK. In order to identify the gyro decay pattern, trend equations were plotted from the numerical results. It was noted that for 45 there was a good approximation, whereas for the other angles the curve diverged mainly near the entrance, having a better convergence from 40D, where the rotation intensity is considered weak (S<0.5). According to the literature, a slew number greater than 0.5 is considered a flow with strong gyration with the presence of reverse flow regions (Kitoh). Note that up to the 30D section there is the presence of a strong turning number.

## Static pressure radial - Discussion

### The presence of rotation in the flow induces the formation of a pressure difference between the wall and the axis of symmetry - consequently, the static pressure is forced to increase near the wall, slowing the flow and decreasing near the axis where the flow is accelerated (CRANE and BURLEY, 1976). The radial pressure distribution is an interesting way to visualize this behavior, as shown below. Radial pressure contributes to the trajectory of a particle, since the pressure difference is one of the properties that acts on the flow behavior. The radial pressure gradient increases in the downstream direction, which corresponds to the increase in the tangential velocity gradient, this behavior was also noted by Slot (2010). The difference between the pressure in the center and close to the wall can be associated with the behavior of the axial velocity distribution.
### Downstream of the slewing generator, the pressure difference between the wall and the shaft reduces as the flow develops. This effect counteracts the tendency to increase the static pressure in the axis, which can delay the flow and vortex breakdown, occurs, mainly for turbulent flow (CRANE and BURLEY, 1976).

```python
plot=[1,2,3,4]
distances = [5,20,40,60]
plotPressureradial('estatica', distances)
#plt.savefig('graficos_v2/pressure_estatica_radial.png')
```

## Gradient Velocity Tangencial - Code

```python
def velDivveltang(df, distances):
  plt.figure(figsize=(12, 10))
  plt.subplot(3,1,i)
  for d in distances: 
    vel_tang = df['dados_radial']['vel_tang_'+str(int(d))+'d']
    r_d = df['dados_radial']['r_'+str(int(d))+'d']
    div_v=np.diff(vel_tang)/np.diff(r_d)
    sns.color_palette("husl", 9)
    plt.scatter(str(int(d)), div_v.mean(), s=80, marker='s', label=str(int(d))+'D')

    plt.xlabel("x/D", fontdict={'size': 16})
    plt.ylabel("$\partial \\bar{w} /\partial r$", fontdict={'size': 16})
    plt.grid(True)
    plt.legend(loc="best", prop={'size': 12})
    plt.show; 
```

## Axial Velocity

### The figure below shows the axial velocity contours for the three cases, where it is possible to observe three main regions: near the flow inlet, where the axial velocity is maximum; the recirculation region, where there is the formation of a vortex, in which the dotted lines represent the presence of negative axial velocity; and the region where the flow is close to reaching the developed regime, from 2.5 m onwards. The increase in the length of the recirculation region is notable with the increase in the tangential velocity component.
### Rocha, Banwwart and Ganzarolli (2017) in their study Effects of inlet boundary conditions in an axial hydrocyclone observe that there is the presence of vortex breakdown in the flow for S values at the inlet above 2.5. Although for the 45 case of this study the S value is below 2.5, a recirculation region similar to the other cases studied can be noted. Leibovich (1978) describes the vortex breakdown as a change in vortex structure indicated by the characteristics of the ratio between the tangential and axial velocity component. In this sense, it is observed that even with S at the entrance close to 1.8, there is a change in the vortex structures presented by the velocity profiles presented, which confirms the presence of vortex breakdown as described by Leibovich.



## Tangencial Velocity

### The figure below shows the tangential velocity contours for the three cases. Note that the greater the tangential velocity field, the greater the recirculation region, delimited between the symmetry axis, the tip of the tail and the rotation field. It is possible to associate that this recirculation region influences the total length of the vortex similar to Rankine's vortex. It is observed that the formation of forced rotation of rigid body starts closer to the symmetry axis for the 45 case.



## Stream Line
